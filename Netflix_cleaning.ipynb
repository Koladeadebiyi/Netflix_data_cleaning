#### 3. Dealing With Missing Values

```python
df.isna().sum()
```

The director, cast and country has many nulls while the date_added, rating and duration has few missing values.

Because they are vast, we can't just drop the nan rows so we will have to isolate each column to understand the reason behind the missing values.


```python
# Director column- This is a categorical column, we can fill the missing values with 'Unknown' and this will not affect our analysis.

df.fillna({'director':'unknown'}, inplace = True)
```

```python
# Cast column- This is a categorical column, we can fill the missing values with 'Unknown' and this will not affect our analysis.

df.fillna({'cast': 'unknown'}, inplace = True)
```

```python
# Country column- We will fill the missing values with 'Unknown' and split the messy data and pick the first country in the list.
df.fillna({'country': 'unknown'}, inplace = True)

df['country'] = df['country'].str.split(',').str[0]  # split the country column by comma and take the first country in the list

```

```python
# Date added column- We have 10 null values out of the 8007 rows, we can drop these rows without affecting our analysis.
df = df.dropna(subset=['date_added'])
```

```python
df.isna().sum()
```

```python

df['rating'] = df['rating'].fillna('NR')
```

```python
# Convert the date_added column to datetime format

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
## The duration column is storing two different meanings; Movies -> minutes, TV Shows -> seasons
df['duration'].unique()
```

 We can split the duration column into two separate columns: duration_int and duration_type. The duration_int column will store the integer value of the duration (96), while the duration_type column will store the type of duration (minutes or seasons).

```python

df[['duration_int', 'duration_type']] = df['duration'].str.split(' ', expand=True)
```

```python
# Dropping the duration nan
df = df.dropna(subset=['duration_int'])
```

```python
df['duration_int'] = df['duration_int'].astype(int)
```

```python
# After extracting the duration_int and duration_type columns, we can drop the original duration column as it is no longer needed.

df = df.drop(columns=['duration'])
```
